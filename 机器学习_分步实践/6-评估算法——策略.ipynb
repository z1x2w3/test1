# 分离训练数据集和评估数据集

```python
import warnings
warnings.filterwarnings('ignore')

from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import time
# 导入数据
filename = 'pima_data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename, names=names)
# 将数据分为输入数据和输出结果
array = data.values
X = array[:, 0:8]
Y = array[:, 8]
test_size = 0.33
seed = 4

start = time.time()
X_train, X_test, Y_traing, Y_test = train_test_split(X, Y, test_size = test_size, random_state = seed)
model = LogisticRegression()
model.fit(X_train, Y_traing)
result = model.score(X_test, Y_test)
finish = time.time()

print("算法评估结果：%.3f%%" % (result * 100))
print("耗时(秒)：", finish - start)
```

# K折交叉验证分离

```python
import warnings
warnings.filterwarnings('ignore')

from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import time
# 导入数据
filename = 'pima_data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename, names = names)
# 将数据分为输入数据和输出结果
array = data.values
X = array[:, 0:8]
Y = array[:, 8]
num_folds = 10
seed = 7

start = time.time()
kfold = KFold(n_splits=num_folds, shuffle = True, random_state = seed) # 没有shuffle=True会提示错误
model = LogisticRegression()
result = cross_val_score(model, X, Y, cv = kfold)
finish = time.time()

print("算法评估结果：%.3f%% (%.3f%%)" % (result.mean() * 100, result.std() * 100))
print("耗时(秒)：", finish - start)
```

# 弃一交叉验证分离

```python
import warnings
warnings.filterwarnings('ignore')

from pandas import read_csv
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import time
# 导入数据
filename = 'pima_data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename, names=names)
# 将数据分为输入数据和输出结果
array = data.values
X = array[:, 0:8]
Y = array[:, 8]

start = time.time()
loocv = LeaveOneOut()
model = LogisticRegression()
result = cross_val_score(model, X, Y, cv = loocv)
finish = time.time()

print("算法评估结果：%.3f%% (%.3f%%)" % (result.mean() * 100, result.std() * 100))
# 注意，这个适用于小样本的计算，比较耗时
print("耗时(秒)：", finish - start)
```

# 重复随机分离评估数据集与训练数据集

```python
from pandas import read_csv
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import time
# 导入数据
filename = 'pima_data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename, names=names)
# 将数据分为输入数据和输出结果
array = data.values
X = array[:, 0:8]
Y = array[:, 8]
n_splits = 10
test_size = 0.33
seed = 7

start = time.time()
kfold = ShuffleSplit(n_splits = n_splits, test_size = test_size, random_state = seed)
model = LogisticRegression()
result = cross_val_score(model, X, Y, cv = kfold)
finish = time.time()

print("算法评估结果：%.3f%% (%.3f%%)" % (result.mean() * 100, result.std() * 100))
print("耗时(秒)：", finish - start)
```
