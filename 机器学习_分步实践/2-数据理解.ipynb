# 描述性统计

```python
from pandas import read_csv
from pandas import set_option
# 描述性统计
filename = 'pima_data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename, names=names)

set_option('display.width', 100)
# 设置数据的精确度
set_option('display.precision', 4)  # 这里原来是'precision'，改成'display.precision'

print(data.describe())
```

# 数据属性的相关性

```python
from pandas import read_csv
from pandas import set_option
# 显示数据的相关性
filename = 'pima_data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename, names=names)

set_option('display.width', 100)
# 设置数据的精确度
set_option('display.precision', 2)  # 这里原来是'precision'，改成'display.precision'

print(data.corr(method='pearson')) # 用pearson相关系数计算属性相关性
```

# 数据的分布分析

```python
from pandas import read_csv
# 计算数据的高斯偏离
filename = 'pima_data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename, names=names)

print(data.skew()) # 偏度
print(data.kurtosis()) # 峰度
```
