# 网格搜索优化算法
GridSearchCV

```python
from pandas import read_csv
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
# 导入数据
filename = 'pima_data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename, names=names)
# 将数据分为输入数据和输出结果
array = data.values
X = array[:, 0:8]
Y = array[:, 8]
# 算法实例化
model = Ridge()

# 设置要遍历的参数
param_grid = {'alpha': [1, 0.1, 0.01, 0.001, 0]} # 超参数网格
# 通过网格搜索查询最优参数
grid = GridSearchCV(estimator = model, param_grid = param_grid)
grid.fit(X, Y)

# 搜索结果
print('最高得分：%.3f' % grid.best_score_)
print('最优参数：%s' % grid.best_estimator_.alpha)
```

```python
from pandas import read_csv
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# 导入数据
filename = 'pima_data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename, names=names)

# 将数据分为输入数据和输出结果
array = data.values
X = array[:, 0:8]
y = array[:, 8]  # 注意：将Y改为y以符合命名惯例


# 创建带预处理和分类器的管道
model = make_pipeline(
    StandardScaler(),  # 标准化数据（提高模型性能）
    LogisticRegression(penalty='l2', solver='liblinear', max_iter=1000)  # 可解释的L2正则化
)

# 设置要遍历的参数
param_grid = {
    'logisticregression__C': [100, 10, 1.0, 0.1, 0.01]  # C是正则化强度的倒数
}


# 使用分层K折交叉验证（保持类别分布）
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 通过网格搜索查询最优参数
grid = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    cv=cv,
    scoring='accuracy'  # 明确使用分类准确率作为评分标准
)
grid.fit(X, y)




# 搜索结果
print('最高准确率：%.3f' % grid.best_score_)
print('最优参数(C)：%.4f' % grid.best_estimator_.named_steps['logisticregression'].C)
```
