# 多重图表

```python
# 相关矩阵图
from pandas import read_csv
import matplotlib.pyplot as plt
import numpy as np
filename = 'pima_data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename, names=names)


correlations = data.corr()
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0, 9, 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

```python
# 散点矩阵图
from pandas import read_csv
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
filename = 'pima_data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename, names=names)

scatter_matrix(data, figsize=(12,10))
plt.show()
```

# 数据可视化总结

```python
# 直方图 Histogram
from pandas import read_csv
import matplotlib.pyplot as plt
filename = 'pima_data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename, names=names)


data.hist(figsize=(12, 10))
# plt.savefig("hist.png") # 保存图片

data.plot(kind='density', subplots=True, figsize=(12, 10), layout=(3,3), sharex=False)
# plt.savefig("density.png") # 保存图片

data.plot(kind='box', subplots=True, figsize=(12, 10), layout=(3,3), sharex=False)
# plt.savefig("box.png") # 保存图片

correlations = data.corr()
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0, 9, 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.savefig("correlations.png") # 保存图片

scatter_matrix(data, figsize=(12,10))
# plt.savefig("scatter_matrix.png") # 保存图片
plt.show()
```
